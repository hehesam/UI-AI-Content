# Seaborn

```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset("tips") #load dataset
sns.relplot(x="total_bill", y="tip", data=tips)
```

```python
sns.relplot(x="total_bill", y="tip", hue="smoker", data=tips)
```

```python
sns.relplot(x="total_bill", y="tip", size="size", data=tips)
```

```python
sns.relplot(x="total_bill", y="tip", size="size", sizes=(15, 200), data=tips)
```

```python
df = pd.DataFrame(dict(time=np.arange(500), value=np.random.randn(500).cumsum()))
g = sns.relplot(x="time", y="value", kind="line", data=df)
g.figure.autofmt_xdate()
```

```python
df = pd.DataFrame(np.random.randn(500, 2).cumsum(axis=0), columns=["x", "y"])
sns.relplot(x="x", y="y", sort=False, kind="line", data=df)
```

```python
fmri = sns.load_dataset("fmri") #load dataset
sns.relplot(x="timepoint", y="signal", hue="region", style="event", dashes=False, markers=True, kind="line", data=fmri)
```

```python
sns.relplot(x="timepoint", y="signal", hue="subject",col="region", row="event", height=3, kind="line", estimator=None, data=fmri)
```

```python
sns.relplot(x="timepoint", y="signal", hue="event", style="event", col="subject", col_wrap=5, height=3, aspect=.75, linewidth=2.5, kind="line", data=fmri.query("region == 'frontal'"))
```

```python
dots = sns.load_dataset("dots").query("align == 'dots'")
palette = sns.cubehelix_palette(light=.8, n_colors=6)
sns.relplot(x="time", y="firing_rate", hue="coherence", style="choice", palette=palette, kind="line", data=dots)
```

```python
sns.displot(penguins, x="flipper_length_mm", hue="sex")
```

```python
sns.displot(penguins, x="flipper_length_mm", col="sex")
```

```python
sns.displot(penguins, x="flipper_length_mm", kind="kde")
```

```python
sns.displot(penguins,x="flipper_length_mm",hue="sex",kind="kde",fill=True)
```

```python
sns.displot(penguins, x="flipper_length_mm", kind="ecdf")
```

```python
sns.displot(penguins, x="flipper_length_mm", hue="species", kind="ecdf")
```

```python
sns.pairplot(penguins)
```

```python
g = sns.PairGrid(penguins)
g.map_upper(sns.histplot)
g.map_lower(sns.kdeplot, fill=True)
g.map_diag(sns.histplot, kde=True)
```

```python
pokemon = pd.read_csv("pokemon.csv") #load dataset
plt.figure(figsize=(10,6))
sns.catplot(y = "Type 1", kind = "count",data = pokemon)
plt.show()
```

```python
plt.figure(figsize=(20,20))
sns.catplot(x='Generation' , y='Defense' , data=pokemon , col="Legendary")
```

```python
sns.catplot(x="Generation", y="Defense", jitter=False, data=pokemon)
```

```python
tips = pd.read_csv("tips.csv") #load dataset
sns.catplot(x="total_bill", y="day", hue="time", kind="swarm", data=tips)
```

```python
sns.catplot(x="day", y="total_bill", kind="box", data=tips)
```

```python
diamonds = sns.load_dataset("diamonds") #load dataset
sns.catplot(x="color", y="price", kind="boxen",
data=diamonds.sort_values("color"))
```

```python
plt.figure(figsize=(7,7))
sns.catplot(x="Generation", y = "Speed", kind="violin", data=pokemon)
plt.show()
```

```python
sns.catplot(x="day", y="total_bill", hue="sex", kind="violin", split=True, data=tips)
```

```python
g = sns.catplot(x="day", y="total_bill", kind="violin", inner=None, data=tips)
sns.swarmplot(x="day", y="total_bill", color="k", size=3, data=tips, ax=g.ax)
```

```python
plt.figure(figsize=(20,7))
sns.barplot(x=pokemon['Type 1'], y= pokemon['Attack'])
plt.show()
```

```python
plt.figure(figsize=(20,7))
order = pokemon.groupby(['Type 1']).mean().sort_values('Defense' , ascending = False).index
sns.barplot(x=pokemon['Type 1'], y= pokemon['Defense'] , order=order)
plt.show()
```
